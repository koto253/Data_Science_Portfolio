Noticed that the `glucose`,    `bloodpressure` `skinthickness`, `insulin`, and `BMI` are all 0
How can Blood Pressure or BMI be 0? That makes no sense! These values also seem to be way off from their respective medians and means, another indicator that something is of

```python
# replacing of 0 with NaN in the `five` columns mentioned 
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
df.describe()
```

```python
# checking for missing values
df.isna().sum()
```

```python
# printing out all of the rows with missing data
df[df.isna().any(axis=1)]
```

```python
# Fill NaN values with the mean of each column
df.fillna(df.median(), inplace=True)
```
